```python
X=wine.drop(columns=['quality'],axis=1)
```

```python
X
```

```python
y = wine['quality']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaler_X_train=scaler.transform(X_train)
```

```python
scaler_X_test = scaler.transform(X_test)
```

```python
print("Distribuição das classes de treino antes do SMOTE")
print(y_train.value_counts())
```

```python
!pip install imbalanced-learn -q
```

```python
k_neighbors = 3
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42,k_neighbors=k_neighbors)
```

```python
X_train_resampled, y_train_resampled = smote.fit_resample(scaler_X_train, y_train)
```

```python
print("\nDistribuição das classes de treino DEPOIS do SMOTE:")
print(y_train_resampled.value_counts())
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train_resampled,y_train_resampled)
```

```python
y_pred = knn.predict(scaler_X_test)
```

```python
from sklearn.metrics import  classification_report
```

```python
print(knn.score(scaler_X_test,y_test))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn import svm
```

```python
svm_model = svm.SVC(class_weight='balanced')
```

```python
svm_model.fit(X_train_resampled,y_train_resampled)
```

```python
y_pred=svm_model.predict(scaler_X_test)
```

```python
print(classification_report(y_test,y_pred))
```
